## ALL Notes for the lesson
- Lists are a good way to organize and collect data in a program, as they allow a set of potentially very different items to be found within one object.
- They are a collection of data 
- Items in lists are located with indexes, which, in most code languages, start with 0 (though the pseudocode used by CollegeBoard usually starts at 1!).
- Using functions like append(), pop(), and insert(), you can add new things to a list or remove them.
- Various inbuilt functions help with performing taskes on lists. .append() and others
- Keeping items within lists allow you to apply a function to each item of a list using iteration.
- Iteration is the repetition of a function. Loops can be used to do this
- for loops - These loops are good for applying a certain alogrithm or function to an entire list of similar things. However, how many times the function iterates is not as controllable.
- Recursive loops can be created in multiple different ways, and involve incrementing a variable until it reaches a certain break point. 
- While loops don't require a function that is then called again within the original function until a condition is met.
