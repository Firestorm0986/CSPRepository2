```python
%%HTML
<html>
<head>
  <title>Car Likes Table</title>
</head>
<body>

<table id="carTable">
  <thead>
    <tr>
      <th>Name of the Car</th>
      <th>Number of Likes</th>
      <th></th>
    </tr>
  </thead>
  <tbody>
    <!-- Table body will be dynamically generated -->
  </tbody>
</table>

<script>
  // Define Car class
  class Car {
    constructor(name) {
      this.name = name;
      this.likes = 0;
    }

    increaseLikes() {
      this.likes++;
    }
  }

  // Create car objects
  const car1 = new Car("Toyota Camry");
  const car2 = new Car("Honda Civic");
  const car3 = new Car("Ford Mustang");

  // Array to hold car objects
  const cars = [car1, car2, car3];

  // Get table body element
  const tableBody = document.querySelector("#carTable tbody");

  // Generate table rows dynamically
  cars.forEach(car => {
    const row = document.createElement("tr");
    
    const nameCell = document.createElement("td");
    nameCell.textContent = car.name;

    const likesCell = document.createElement("td");
    likesCell.textContent = car.likes;

    const buttonCell = document.createElement("td");
    const likeButton = document.createElement("button");
    likeButton.textContent = "Like";
    likeButton.addEventListener("click", () => {
      car.increaseLikes();
      likesCell.textContent = car.likes;
    });

    buttonCell.appendChild(likeButton);

    row.appendChild(nameCell);
    row.appendChild(likesCell);
    row.appendChild(buttonCell);

    tableBody.appendChild(row);
  });
</script>

</body>
</html>
```
